## **4. Insigths**

Only count people's orginal's cz and destination's cz are different.

```python
od_race['condition'] = np.where((od_race['o_cz'] == od_race['d_cz']), 'remove', 'keep')
display(od_race)
```

```python
sorted = od_race.sort_values(by = ['condition'], ascending = False)
sorted.head(10)
```

```python
df1 = od_race.groupby(['o_state_name','pool','condition']).agg({'n':['mean', 'sum']})
df1.columns = ['n_mean', 'n_sum']
df1 = df1.reset_index()
print(df1)
```

```python
df1_cz = od_race.groupby(['o_state_name','o_cz_name','pool','condition']).agg({'n': ['sum']})
```

### **Data Sanity Check**

```python
df2 = df1.query("condition == 'keep'")
df2_sorted = df2.sort_values(by=['n_sum','o_state_name'], ascending = [False,True])
print(df2)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

plot_df2 = df2_sorted.pivot_table(index = "o_state_name", columns = "pool", values = "n_sum")

plot_df2 \
    .plot(kind = "bar", stacked = True, figsize = (30, 8), rot = 0) \
    .set(xlabel = "original state",
         ylabel = "num of migration",
         title =  "Distribution of Migration by States & Race")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
