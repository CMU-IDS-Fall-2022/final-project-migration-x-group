```python
#Get all the Dataframes
dataframe_array = [df2013_subset, df2014_subset, df2015_subset, df2016_subset, df2017_subset, 
                   df2018_subset]
```

```python
for frame in dataframe_array:
    print(frame)
```

```python
#Get the substring of the state
new_frame_array = []
for frame in  dataframe_array:
    state = []
    for name in frame['Label']:
        start = name.find(',')+2
        name = name[start:]
        state.append(name)
        
    frame['STATE'] = state
    df_state=frame[["STATE", "Educational ratio"]]
    new_frame_array.append(df_state)
```

```python
new_frame_array
```

```python
df_2013 = new_frame_array[0]
print(df_2013)
```

```python
df_2013.dtypes
```

```python
#Get the average of all the states
states = new_frame_array[0]['STATE'].unique()
print(states)
```

```python
states.shape
```

```python
mean_of_2013_2018 = []
for frame in new_frame_array:
    mean = []
    for state in states:
        fil = frame["STATE"] == state
        select = frame[fil]
        select_mean = select['Educational ratio'].mean()
        mean.append(select_mean)
    mean_of_2013_2018.append(mean)

for m in mean_of_2013_2018:
    print(m)
```

```python
#Get the average of all the year for each individual state
average = []

for i in range(0, 52, 1):
    sum = 0
    for m in mean_of_2013_2018:
        sum += m[i]
    aver = sum / 11
    average.append(aver)
    
print(average) 
```

```python
average_data ={"state": states,
       "average": average}
final = pd.DataFrame(average_data)
print(final)
```

```python
final.to_csv('data/EducationalByState.csv')
```
