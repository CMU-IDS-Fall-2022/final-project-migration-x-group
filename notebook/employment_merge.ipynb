# Step1: #Drop the first row and first column

```python
new_frame_array = []
for frame in dataframe_array:
    frame = frame.iloc[1: , :]
    frame =  frame.iloc[: , 1:]
    new_frame_array.append(frame)

for frame in  new_frame_array:
    print(frame)
```

# Step 2: get the substring of the state


```python
new_frame_array1 = []
for frame in  new_frame_array:
    state = []
    for name in frame['NAME']:
        start = name.find(',')+2
        name = name[start:]
        state.append(name)
        
    frame['STATE'] = state
    df_state=frame[["STATE", "DP03_0004PE"]]
    new_frame_array1.append(df_state)

        
# for frame in new_frame_array1:
#     print(frame)
```

```python
df_2011 = new_frame_array1[0]
print(df_2011)
```

# Step3: convert to proper data types

```python
df_converted = []
```

```python
for frame in new_frame_array1:
    frame = frame.astype(dtype={'STATE': 'string'})
    frame['DP03_0004PE'].replace(to_replace ="N", value = 50, inplace = True)
    frame = frame.astype(dtype={'DP03_0004PE': 'float'})
    df_converted.append(frame)

for f in df_converted:
    print(f.dtypes)
    
```

# Step4: get the average of all the states

```python
states = new_frame_array1[0]['STATE'].unique()
print(states)
```

```python
mean_of_2011_2021 = []
for frame in df_converted:
    mean = []
    for state in states:
        fil = frame["STATE"] == state
        select = frame[fil]
        select_mean = select['DP03_0004PE'].mean()
        mean.append(select_mean)
    mean_of_2011_2021.append(mean)

for m in mean_of_2011_2021:
    print(m)
```

# Step 4: Get the average of all the year for each individual state

```python
average = []

for i in range(0, 52, 1):
    sum = 0
    for m in mean_of_2011_2021:
        sum += m[i]
    aver = sum / 11
    average.append(aver)
    
print(average) 
```

```python
average_data ={"state": states,
       "average": average}
final = pd.DataFrame(average_data)
print(final)
```

```python
final.to_csv(index=False)
compression_opts = dict(method='zip',
                        archive_name='out.csv')  
final.to_csv('out.zip', index=False,
          compression=compression_opts)  

from pathlib import Path  
filepath = Path('/Users/christy/Desktop/out.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
final.to_csv(filepath)

import os  
os.makedirs('folder/subfolder', exist_ok=True)  
final.to_csv('/Users/christy/Desktop/out.csv')  
```
