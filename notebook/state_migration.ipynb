```python
import pandas as pd
```

```python
state_migration = pd.read_csv("../data/state_migration.csv")
state_migration
```

Group by origin state
calculate the sum of n


```python
o_state_groups = state_migration.groupby("o_state_name")
```

```python
o_state_groups.get_group("Alabama")
```

```python
filtered_data = o_state_groups.get_group("Alabama")
filtered_data["%_of_total"] = (filtered_data["n"] / filtered_data["n"].sum()) * 100
filtered_data
```
